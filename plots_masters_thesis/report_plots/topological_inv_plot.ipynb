simple script to make a paper figure.

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the plot
x = np.linspace(0, 10, 100)  # X-axis values
x_c = 5  # Critical parameter where the step function changes
y_line = x  # Straight line: y = x
y_step = np.where(x < x_c, 0, 1)  # Step function: 0 before x_c, 1 after x_c

# Create the figure and the first y-axis
fig, ax1 = plt.subplots()

# Plot the y = x line on the left y-axis
color = 'blue'
ax1.plot(x, y_line, color=color, label="y = x")
ax1.set_xlabel('Parameter Value (x)')
ax1.set_ylabel('y-axis: $y_0$ to $y_1$', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create the second y-axis for the step function
ax2 = ax1.twinx()  # Shared x-axis
color = 'red'
ax2.step(x, y_step, color=color, where='mid', label="Step function")
ax2.set_ylabel('Topological Invariant: $t_0$ to $t_1$', color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Add grid and a legend
ax1.grid(alpha=0.3)
fig.tight_layout()  # Adjust layout to prevent overlap

# Show the plot
plt.title("Dual Y-Axis Plot with Critical Parameter $x_c$")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the plot
x = np.linspace(0, 10, 100)  # X-axis values
x_c = 5  # Critical parameter where the step function changes
y_line = x  # Straight line: y = x
y_step = np.where(x < x_c, 0, 1)  # Step function: 0 before x_c, 1 after x_c

# Create the figure and the first y-axis
fig, ax1 = plt.subplots()

# Plot the y = x line on the left y-axis
color = 'C0'#'blue'
ax1.plot(x, y_line, color=color, label="y = x")
ax1.set_xlabel('Parameter Value')
ax1.set_ylabel('Continuous System Property', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Set custom ticks and labels for the x-axis and left y-axis
ax1.set_xticks([x[0], x[len(x)//2], x[-1]])
ax1.set_xticklabels(['$p_0$', '$p_c$', '$p_1$'])
ax1.set_yticks([y_line[0], y_line[-1]])
ax1.set_yticklabels(['$H(y_0)$', '$H(y_1)$'])

# Create the second y-axis for the step function
ax2 = ax1.twinx()  # Shared x-axis
color = 'C1'#'red'
ax2.step(x, y_step, color=color, where='mid', label="Step function")
ax2.set_ylabel('Topological Invariant', color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Set custom ticks and labels for the right y-axis
ax2.set_yticks([0, 1])  # Min and max values of the step function
ax2.set_yticklabels(['$t_0$', '$t_1$'])

# Add grid and a legend
# ax1.grid(alpha=0.3)
ax1.grid(False)
fig.tight_layout()  # Adjust layout to prevent overlap

# Show the plot
# plt.title("Dual Y-Axis Plot with Critical Parameter $x_c$")
# plt.show()
plt.savefig('/Users/Will/Documents/FINAL_PROJECT/simulations/report_plots/abstract_invariant.png', dpi=1000)
```
